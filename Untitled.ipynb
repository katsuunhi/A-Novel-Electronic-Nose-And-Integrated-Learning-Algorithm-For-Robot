```python
#分类算法的正确性评判
cmap = sn.cubehelix_palette(start = 1.5, rot = 3, gamma=0.8, as_cmap = True)
def fit_models(name, model, x_train, y_train, x_test, y_test):
    model.fit(x_train,y_train)
    y_pred = model.predict(x_test)
    print("{}训练集准确率：{}".format(name,accuracy_score(y_train,model.predict(x_train))))
    print("测试集准确率：",accuracy_score(y_test,y_pred))
    sn.heatmap(confusion_matrix(y_test, y_pred), annot = True, cmap = plt.cm.Blues)  #混淆矩阵
    return accuracy_score(y_test,y_pred)
```

```python
#回归算法的正确性评判
def r_fit_models(name, model, x_train, y_train, x_test, y_test):
    model.fit(x_train,y_train)
    test_predict = model.predict(x_test)
    for i in range(len(test_predict)):
        test_predict[i] = round(test_predict[i])
    print("{}测试集准确率：{}".format(name, accuracy_score(y_test, test_predict)))
    sn.heatmap(confusion_matrix(y_test, test_predict), annot = True, cmap=plt.cm.Blues)#cmap)  #混淆矩阵
    return accuracy_score(y_test,test_predict)
```

```python
def rmse_cv(model,X,y):
    rmse = np.sqrt(-cross_val_score(model, X, y, scoring="neg_mean_squared_error", cv=5))
    return rmse
```

```python
#回归模型的集成
class AverageWeight(BaseEstimator, RegressorMixin):
    def __init__(self,mod,weight):
        self.mod = mod
        self.weight = weight
        
    def fit(self,X,y):
        self.models_ = [clone(x) for x in self.mod]
        for model in self.models_:
            model.fit(X,y)
        return self
    def predict(self,X):
        w = list()
        pred = np.array([model.predict(X) for model in self.models_])
        for data in range(pred.shape[1]):
            single = [pred[model,data]*weight for model,weight in zip(range(pred.shape[0]),self.weight)]
            w.append(np.sum(single))
        return w
```

```python
#回归算法
r_names = ["RFR", "GBR", "ETR", "LGR", "XGR", "SVM"]
r_models = [RandomForestRegressor(), GradientBoostingRegressor(), ExtraTreesRegressor(),
            lgb.LGBMRegressor(),XGBRegressor(),svm.SVC(C=1.0,kernel='linear')]
GBR = GradientBoostingRegressor()
ETR = ExtraTreesRegressor()
LGR = lgb.LGBMRegressor()
XGR = XGBRegressor()


#分类器
names = ["RF", "GBC","DTC","KNN","LGB","XGB","SVM","GNB","LR","GPC"]
models = [RandomForestClassifier(10),GradientBoostingClassifier(),DecisionTreeClassifier(),KNeighborsClassifier(10),
          lgb.LGBMClassifier(),XGBClassifier(),svm.SVC(C=1.0,kernel='linear'),GaussianNB(),LogisticRegression(), GaussianProcessClassifier()]
SVM = svm.SVC(C=1.0, kernel='linear', decision_function_shape='ovr',probability=True)
KNN = KNeighborsClassifier(10)
GPC = GaussianProcessClassifier()
GNB = GaussianNB()
DTC = DecisionTreeClassifier()
```

```python
#回归模型求最优权重
best = [0,0,0,0]
rate = 5
for w1 in range(20, 40, rate):
    for w2 in range(0, 100 - w1, rate):
        w3 = 100 - w1 - w2
        weight_avg = AverageWeight(mod = [SVM, XGR, LGR],weight=[w1/100.0,w2/100.0,w3/100.0])
        score = r_fit_models('weight', weight_avg, x_train, y_train, x_test, y_test)
        if score > best[3]:
            best[0] = w1
            best[1] = w2
            best[2] = w3
            best[3] = score
            print(best)
```
